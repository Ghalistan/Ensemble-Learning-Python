### Import Library

```python
from sklearn.naive_bayes import GaussianNB
import numpy as np
import matplotlib.pyplot as plt
```

### Get data

```python
trainset = np.genfromtxt('TrainsetTugas4ML.csv', delimiter=',', skip_header=1)
testset = np.genfromtxt('TestsetTugas4ML.csv', delimiter=',', skip_header=1)
```

### Bagging and GaussianNB
implementing Gaussian Naive Bayes and Bagging

```python
bagging = 25
Models = np.zeros(shape=(0,2))
for i in range(bagging):
    get = np.random.choice(len(trainset), 50)
    subdataset = trainset[get]
    NB = GaussianNB(var_smoothing=0.2).fit(subdataset[:,:1], subdataset[:,2])
    pred = NB.predict(subdataset[:,:1])
    mean = np.mean(np.equal(subdataset[:,2], pred))
    Models = np.append(Models, [[NB, mean]], axis=0)
    
print("Models Average Accuracy to itself : {}".format(np.mean(Models[:,1])))
print(Models)
```

### Calculate Model Accuracy
calculate Model accuracy to current trainset

```python
predict = []
for m in Models:
    predict.append(m[0].predict(trainset[:,:1]))

classes = np.zeros(0).astype(int)
for i in range(len(trainset)):
    classes = np.append(classes, np.argmax(np.bincount([p[i] for p in predict])))

print("Akurasi Bagging Naive Bayes Terhadap trainset : {}".format(np.mean(np.equal(trainset[:,2], classes))*100))
```

### Trainset visualization

```python
class1 = trainset[np.where(trainset[:,2] == 1)]
class2 = trainset[np.where(trainset[:,2] == 2)]
plt.scatter(class1[:,0], class1[:,1], label='class = 1')
plt.scatter(class2[:,0], class2[:,1], label='class = 2')
plt.legend()
plt.show()
```

### Trainset with Classes prediction visualization

```python
class1 = trainset[np.where(classes == 1)]
class2 = trainset[np.where(classes == 2)]
plt.scatter(class1[:,0], class1[:,1], label='class = 1')
plt.scatter(class2[:,0], class2[:,1], label='class = 2')
plt.legend()
plt.show()
```

### Applying Models to actual testset

```python
predict = []
for m in Models:
    predict.append(m[0].predict(testset[:,:1]))

classes = np.zeros(0).astype(int)
for i in range(len(testset)):
    classes = np.append(classes, np.argmax(np.bincount([p[i] for p in predict])))
```

### Save to CSV

```python
np.savetxt("TebakanTugas4ML.csv", classes.astype(int), fmt="%i", delimiter=",")
```

### Testset and Class prediction Visualized

```python
class1 = testset[np.where(classes == 1)]
class2 = testset[np.where(classes == 2)]
plt.scatter(class1[:,0], class1[:,1], label='Class = 1')
plt.scatter(class2[:,0], class2[:,1], label='Class = 2')
plt.legend()
plt.show()
```
